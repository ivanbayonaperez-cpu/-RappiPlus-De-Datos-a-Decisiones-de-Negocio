### Validar y convertir fechas

```python
# Convertir fecha_hora_pedido de texto a datetime real
orders['fecha_hora_pedido'] = pd.to_datetime(orders['fecha_hora_pedido'], errors='coerce')
```

```python
# Verificar si quedó algún nulo tras la conversión (fechas mal escritas)
print(orders['fecha_hora_pedido'].isnull().sum())
```

```python

# Revisar el rango de fechas para detectar algo fuera de lo esperado
print(orders['fecha_hora_pedido'].min(), '→', orders['fecha_hora_pedido'].max())
```

### Revisar variables numéricas (negativos / ceros inválidos)

```python
columnas_numericas = ['cantidad', 'precio_unitario', 'monto_descuento', 'monto_total']

# Ver estadísticas generales: mínimo, máximo, promedio
print(orders[columnas_numericas].describe())

# Contar cuántas filas tienen cantidad negativa
print('Cantidad negativa:', (orders['cantidad'] < 0).sum())

# Contar cuántas filas tienen precio_unitario en cero o negativo (inválido, un producto no puede costar $0 o menos)
print('Precio inválido:', (orders['precio_unitario'] <= 0).sum())

# Ver si el descuento es mayor al valor del producto (inconsistente)
print('Descuento mayor al precio:', (orders['monto_descuento'] > orders['precio_unitario']).sum())


# Bandera de devoluciones (cantidad negativa) — se crea AQUÍ, junto con el resto de banderas
orders['es_devolucion'] = orders['cantidad'] < 0
print(orders['es_devolucion'].sum(), 'pedidos marcados como devolución')
```

### Verificar consistencia de montos

```python

# Fórmula corregida desde el inicio (ya sabemos que las devoluciones invierten el signo del descuento)
orders['monto_calculado'] = (orders['precio_unitario'] * orders['cantidad']) - (orders['monto_descuento'] * np.sign(orders['cantidad']))
orders['diferencia'] = (orders['monto_total'] - orders['monto_calculado']).abs()
orders['monto_inconsistente'] = orders['diferencia'] > 0.02

print(orders['monto_inconsistente'].sum(), 'filas realmente inconsistentes')
```

### Eliminar duplicados

```python

for nombre, df in [('orders', orders), ('catalog', catalog), ('marketing', marketing)]:
    exactos = df.duplicated().sum()
    print(f'{nombre}: {exactos} filas exactamente duplicadas')

duplicados_id = orders['id_pedido'].duplicated().sum()
print(f'id_pedido repetidos: {duplicados_id}')

orders = orders.drop_duplicates()
catalog = catalog.drop_duplicates()
marketing = marketing.drop_duplicates()
```

*¿Cuál es el ingreso total (revenue)?*

```python
revenue_total = orders['monto_total'].sum()
print(f'Revenue total: ${revenue_total:,.2f}')
```

*¿Cuál es el costo total?*

```python
# Unir orders con catalog, usando nombre_producto como llave de conexión
orders_con_costo = orders.merge(
    catalog[['nombre_producto', 'costo_unitario']],
    on='nombre_producto',
    how='left'
)

# Calcular el costo de cada pedido: costo unitario x cantidad comprada
orders_con_costo['costo_pedido'] = orders_con_costo['costo_unitario'] * orders_con_costo['cantidad']

costo_total = orders_con_costo['costo_pedido'].sum()
print(f'Costo total: ${costo_total:,.2f}')
```

```python
# verificacion 
print(orders_con_costo['costo_unitario'].isnull().sum(), 'pedidos sin costo calculado (por nombre_producto faltante)')
```

```python
# margen bruto 
margen_bruto_monto = revenue_total - costo_total
print(f"Margen Bruto ($): ${margen_bruto_monto:,.2f}")
```

*¿Cuánto se ha invertido en marketing?*

```python
marketing_total = marketing['gasto'].sum()
print(f'Inversión total en marketing: ${marketing_total:,.2f}')
```

```python
# Verificacion que no quedo ningun nulo 
print(marketing['gasto'].isnull().sum(), 'valores nulos en gasto')
```

```python
# calculo de ¿El negocio es rentable? (Profit)
profit_total = revenue_total - costo_total - marketing_total
print(f'Profit total: ${profit_total:,.2f}')

# Margen de profit como porcentaje del revenue (más fácil de interpretar que el número absoluto)
margen_profit = (profit_total / revenue_total) * 100
print(f'Margen de profit: {margen_profit:.1f}%')
```

```python
# ¿Cuánto pesan los outliers en el revenue y costo total?
revenue_outliers = orders[orders['outlier_cantidad']]['monto_total'].sum()
costo_outliers = orders_con_costo[orders_con_costo['outlier_cantidad']]['costo_pedido'].sum()

print(f'Revenue de pedidos outlier: ${revenue_outliers:,.2f} ({revenue_outliers/revenue_total*100:.1f}% del total)')
print(f'Costo de pedidos outlier: ${costo_outliers:,.2f} ({costo_outliers/costo_total*100:.1f}% del total)')
```

*Piensa en la magnitud de esto: tienes 24,990 pedidos "normales" generando apenas 18.5% del revenue ($51.9M × 0.185 ≈ $9.6M), mientras que 10 pedidos con cantidades de 10,000–20,000 unidades generan el resto. Eso no es un negocio de delivery/suscripción real — ningún usuario de RappiPlus compra 20,000 tablets en un solo pedido. Esto confirma con mucha fuerza la hipótesis que ya habíamos planteado en el Paso 1: son errores de captura (probablemente un par de ceros de más), no ventas reales.
Por qué esto cambia todo tu análisis
Tu profit_total = $5,991,600.32 y margen = 11.5% están calculados sobre datos dominados por un error. Es decir: ese "11.5% de margen" no describe cómo se comporta el negocio real de RappiPlus — describe principalmente 10 filas rotas. Si reportaras este número al jefe tal cual, estarías respondiendo la pregunta de negocio ("¿el negocio es rentable?") con una cifra que en su mayoría no refleja el comportamiento real de los usuarios.
Esto es exactamente el momento en el que, como analista, no avanzas con el número que salió — te detienes, investigas la causa raíz (ya lo hiciste en el Paso 1), y ahora tomas una decisión explícita sobre cómo tratarlo antes de calcular el KPI final.*

```python
# Recalcular revenue, costo y profit EXCLUYENDO los pedidos marcados como outlier
orders_sin_outliers = orders[~orders['outlier_cantidad']]
orders_con_costo_sin_outliers = orders_con_costo[~orders_con_costo['outlier_cantidad']]

revenue_real = orders_sin_outliers['monto_total'].sum()
costo_real = orders_con_costo_sin_outliers['costo_pedido'].sum()
profit_real = revenue_real - costo_real - marketing_total
margen_real = (profit_real / revenue_real) * 100

print(f'Revenue real (sin outliers): ${revenue_real:,.2f}')
print(f'Costo real (sin outliers): ${costo_real:,.2f}')
print(f'Profit real (sin outliers): ${profit_real:,.2f}')
print(f'Margen real: {margen_real:.1f}%')


```

```python
print(f'\n--- Comparación ---')
print(f'Profit CON outliers:  ${profit_total:,.2f} (margen {margen_profit:.1f}%)')
print(f'Profit SIN outliers:  ${profit_real:,.2f} (margen {margen_real:.1f}%)')
```

```python
# que tan rentables son los outliers
profit_outliers = revenue_outliers - costo_outliers
margen_outliers = (profit_outliers / revenue_outliers) * 100
print(f'Profit de los outliers (antes de marketing): ${profit_outliers:,.2f}')
print(f'Margen de los outliers: {margen_outliers:.1f}%')
```

### Paso 1: identificar la cohorte de cada usuario ###

```python
query_cohortes = '''
SELECT 
    id_usuario,
    CAST(fecha_registro AS DATE) AS fecha_registro,
    DATE_TRUNC('month', CAST(fecha_registro AS DATE)) AS mes_cohorte
FROM users;
'''
cohortes = pd.read_sql(query_cohortes, con=engine)
print(cohortes.head())
```

### Paso 2: unir cohortes con actividad, y calcular retención por semana ###

```python
# Retención por cohortes
# ======================


query_cohort_retention_final = '''
WITH cohortes AS (
    SELECT 
        id_usuario,
        DATE_TRUNC('month', CAST(fecha_registro AS DATE)) AS mes_cohorte
    FROM users
),
tamanos_cohorte AS (
    SELECT 
        mes_cohorte,
        COUNT(DISTINCT id_usuario) AS usuarios_iniciales
    FROM cohortes
    GROUP BY mes_cohorte
),
retencion AS (
    SELECT
        c.mes_cohorte,
        COUNT(DISTINCT CASE WHEN ua.dias_despues_registro = 7 AND ua.activo = 1 THEN ua.id_usuario END) AS retenido_w1,
        COUNT(DISTINCT CASE WHEN ua.dias_despues_registro = 14 AND ua.activo = 1 THEN ua.id_usuario END) AS retenido_w2,
        COUNT(DISTINCT CASE WHEN ua.dias_despues_registro = 21 AND ua.activo = 1 THEN ua.id_usuario END) AS retenido_w3
    FROM cohortes c
    JOIN user_activity ua ON c.id_usuario = ua.id_usuario
    GROUP BY c.mes_cohorte
)
SELECT 
    r.mes_cohorte,
    t.usuarios_iniciales,
    r.retenido_w1,
    r.retenido_w2,
    r.retenido_w3,
    ROUND(100.0 * r.retenido_w1 / t.usuarios_iniciales, 1) AS semana_1,
    ROUND(100.0 * r.retenido_w2 / t.usuarios_iniciales, 1) AS semana_2,
    ROUND(100.0 * r.retenido_w3 / t.usuarios_iniciales, 1) AS semana_3
FROM retencion r
JOIN tamanos_cohorte t ON r.mes_cohorte = t.mes_cohorte
ORDER BY r.mes_cohorte;
'''

# Ejecutar la consulta

cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
print(cohorte_final)
```

## 🔹 Paso 5: Validar si los cambios generan impacto (test estadístico)

🎯 **Objetivo:** Evaluar si la modificación en la UI del checkout impacta la **tasa de conversión de compra**.

---

1. **Analizar el dataset** `experiment_checkout_ui.csv` para identificar la métrica principal **conversion**.
   - La métrica **conversion** es 1 si el usuario completó la compra, 0 si no.    
2. **Plantear la hipótesis estadística**     
3. **Aplicar el test estadístico adecuado** 
4. **Interpretar el resultado**  

---
Hipótesis estadística
   - **H₀ (Hipótesis nula):** ...
   - **H₁ (Hipótesis alternativa):** ...
   
**Test estadístico:** ...  
**Nivel de significancia alpha:** ...

```python
import pandas as pd

url_experiment = 'https://practicum-content.s3.amazonaws.com/datasets/experiment_checkout_ui.csv'
experiment = pd.read_csv(url_experiment)

print(experiment.shape)
print(experiment.info())
experiment.head()
```

### 1 paso: verificar el tamaño de cada grupo ###

```python
print(experiment['variante'].value_counts())
```

### 2 paso: Calcular la tasa de conversión por grupo

```python
tasas_conversion = experiment.groupby('variante')['convirtio'].agg(['sum', 'count', 'mean'])
tasas_conversion.columns = ['conversiones', 'total_usuarios', 'tasa_conversion']
print(tasas_conversion)
```

### 3 paso: El test estadístico — prueba de proporciones (z-test) ###

```python
from statsmodels.stats.proportion import proportions_ztest

# Extraemos los valores que necesita la función
conversiones = tasas_conversion['conversiones'].values
totales = tasas_conversion['total_usuarios'].values

z_stat, p_value = proportions_ztest(count=conversiones, nobs=totales)

print(f'Estadístico Z: {z_stat:.4f}')
print(f'P-valor: {p_value:.4f}')
```

### 4 paso: comparar el p-valor contra alpha ###

```python
alpha = 0.05

if p_value < alpha:
    print(f'p-valor ({p_value:.4f}) < alpha ({alpha}) → Se rechaza H0')
    print('Existe evidencia estadísticamente significativa de que el cambio de UI afecta la conversión.')
else:
    print(f'p-valor ({p_value:.4f}) >= alpha ({alpha}) → No se rechaza H0')
    print('No hay evidencia suficiente de que el cambio de UI afecte la conversión.')
```

**Dataset:** `experiment_checkout_ui.csv` — 10,000 usuarios, sin valores nulos
**Grupos:** control (4,965 usuarios) vs. tratamiento (5,035 usuarios) — balanceados correctamente

---

### Hipótesis estadística

- **H₀ (nula):** no existe diferencia real en la tasa de conversión entre el diseño 
  actual del checkout (control) y el nuevo diseño (tratamiento).
- **H₁ (alternativa):** existe una diferencia real en la tasa de conversión entre 
  ambos diseños.

**Test aplicado:** prueba Z de dos proporciones (`proportions_ztest`), adecuada 
para comparar tasas de conversión (variable binaria) entre 2 grupos independientes.
**Nivel de significancia:** alpha = 0.05

---

### Resultado principal

| Grupo | Conversiones | Usuarios | Tasa de conversión |
|---|---|---|---|
| Control | 779 | 4,965 | 15.69% |
| Tratamiento | 820 | 5,035 | 16.29% |

**Estadístico Z:** -0.8133
**P-valor:** 0.4161

Como el p-valor (0.4161) es mayor que alpha (0.05), **no se rechaza H₀**.

**Conclusión:** con los datos disponibles, no hay evidencia estadísticamente 
significativa de que el nuevo diseño de checkout tenga un efecto sobre la tasa 
de conversión. La diferencia observada (+0.6 puntos porcentuales a favor del 
tratamiento) es consistente con variabilidad aleatoria de la muestra, y no 
permite afirmar que el cambio de UI mejora ni empeora la conversión.

*Nota de interpretación: este resultado no demuestra que ambos diseños sean 
"iguales" — solo indica que, con el tamaño de muestra actual (10,000 usuarios), 
no se detectó evidencia suficiente de una diferencia real.*
